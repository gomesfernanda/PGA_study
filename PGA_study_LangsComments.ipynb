### Understand the 10 most popular languages on PGA

```python
langs_list = []
for row in range(len(df)):
    
    row_str = str(df.iloc[row].at['LANGS'])
    row_list = row_str.split(',')
    langs_list += row_list

cnt = Counter()
for word in langs_list:
    cnt[word] += 1
```
